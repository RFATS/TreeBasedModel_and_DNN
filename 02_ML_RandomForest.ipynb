# **1. RandomForest**
**모델구분**  
- **`RandomForest`** 모델은 **supervised** 머신러닝 모델이다.  
- 기본 구조로서 **`DecisionTree`** 를 사용하는 treebase 모델로 prediction, classification이 모두 가능한 모델이다.  

**모델 구조 및 특징**  
- 여러개의 **`DecisionTree`** 모델을 병렬 구성하여 모델을 결합한 **ensemble(앙상블)** 방법론이다.  

**학습**  
- ensemble 방법론 중 **bagging**을 사용하여 **병렬** 구성한 모델의 개별 결과값을 결합하는 방식으로 최종 결과값을 만들어 낸다.

**효과 및 활용**  
- 모델 결과의 분산이 작아져 최종 결과값이 안정적이다.  
- 하이퍼 파라메터 서칭을 활용 해 모델 성능 개선이 가능하다.   
- 모델 구조를 서버단에서 구현하기 좋아 IT업계에서는 오래도록 사용되어 온 모델이다.  

## **2. ensemble(앙상블)**  
- **여러개의 모형을 결합**하여 더 나은 성능을 확보하는 방법론  
- 단일 모형을 사용할 때 보다 **성능 분산이 감소**하고, 즉 **overfitting** 개선하는 효과를 갖는다.  
- 앙상블 방법론에는 **bagging**, **boosting** **stacking** 이 있다.

## **4. RamdomForest 실습**

```python
from sklearn.ensemble import RandomForestRegressor
RandomForestRegressor()
```

```python
# 필요모듈 import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import root_mean_squared_error, r2_score
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# 보스턴 데이터 로드 및 데이터셋 생성
df = pd.read_csv("https://raw.githubusercontent.com/overgroove/ML_lecture_image_data/refs/heads/main/data/boston.csv")
# 타겟데이터 분할
df.head()
# test data 분할
y = df['y']
X = df.drop('y', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
rfr = RandomForestRegressor(random_state=42)
rfr.fit(X_train, y_train)
y_pred = rfr.predict(X_test)
print(r2_score(y_test, y_pred))
print(root_mean_squared_error(y_test, y_pred))
```

```python
'''dtr
R2 score : 0.8593731880046497
RMSE : 3.2370571637853427
'''
```

## **5. 성능 개선을 위해 서칭이 필요한 모델링 파라메터**
- **n_estimators** : 사용 할 트리 모델 갯수  
- **max_depth** : 트리의 최대 깊이
- **max_features** : 개별 트리간 상관관계를 줄여 앙상블 효과를 높임

## **6. 파라메터 서칭**
- tree base 모델은 설정 가능한 파라메터의 조합에 따라 모델 예측력 차이가 큼.  
- 파라메터 설정에 따른 모델 예측력 차이가 크기에 꼭 파라메터 서칭을 진행해주셔야 합니다.  
- 파이썬 기본 문법을 통해 파라메터 서칭을 간단하게 진행 해보겠습니다.

```python
# validation set 생성
X_train_2, X_val, y_train_2, y_val = train_test_split(X_train, y_train, test_size=0.3, random_state=42)
X_train_2.shape, X_val.shape, X_test.shape
```

```python
# 필요모듈 import
from itertools import product
```

```python
# product 함수로 파라메터의 모든 조합 만들기
est = [80, 85, 90, 95]
depth = [15, 16, 17]

best_score = 99999999
best_param = None

for param in product(est, depth):
    print(f"{param} 서칭중")
    model = RandomForestRegressor(n_estimators=param[0], max_depth=param[1], random_state=42)
    model.fit(X_train_2, y_train_2) # 학습데이터는 위에서 분할 할 학습데이터
    pred = model.predict(X_val)
    # 어떤 평가지표를 활용해서 파라메터 서칭에 이용할건가?
    rmse = root_mean_squared_error(y_val, pred)
    # 최소값 찾는 알고리즘
    if rmse < best_score:
        print("스코어 갱신!!!")
        best_score = rmse
        best_param = param

print(f'서칭이 종료 되었습니다. 베스트 파라메터는 {best_param}입니다.')
```

```python
# 최적 r2값, 최적 파라메터 확인
best_param, best_score
```

```python
# 최적 모델로 모델 다시 학습 및 평가
best_model = RandomForestRegressor(n_estimators=best_param[0], max_depth=best_param[1], random_state=42)
best_model.fit(X_train, y_train)
pred = best_model.predict(X_test)
print(r2_score(y_test, pred))
print(root_mean_squared_error(y_test, pred))
```
