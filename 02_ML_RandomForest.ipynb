# **Random Forest**

# **1. RandomForest**
**모델구분**  
- **`RandomForest`** 모델은 **supervised** 머신러닝 모델이다.  
- 기본 구조로서 **`DecisionTree`** 를 사용하는 treebase 모델로 prediction, classification이 모두 가능한 모델이다.  

**모델 구조 및 특징**  
- 여러개의 **`DecisionTree`** 모델을 병렬 구성하여 모델을 결합한 **ensemble(앙상블)** 방법론이다.  

**학습**  
- ensemble 방법론 중 **bagging**을 사용하여 **병렬** 구성한 모델의 개별 결과값을 결합하는 방식으로 최종 결과값을 만들어 낸다.

**효과 및 활용**  
- 모델 결과의 분산이 작아져 최종 결과값이 안정적이다.  
- 하이퍼 파라메터 서칭을 활용 해 모델 성능 개선이 가능하다.   
- 모델 구조를 서버단에서 구현하기 좋아 IT업계에서는 오래도록 사용되어 온 모델이다.  

## **2. ensemble(앙상블)**  
- **여러개의 모형을 결합**하여 더 나은 성능을 확보하는 방법론  
- 단일 모형을 사용할 때 보다 **성능 분산이 감소**하고, 즉 **overfitting** 개선하는 효과를 갖는다.  
- 앙상블 방법론에는 **bagging**, **boosting** **stacking** 이 있다.

## **3. bagging(배깅)**
- **`RandomForest`** 모델이 사용하는 앙상블 방법론으로 **`DecisionTree`** 모델을 병렬로 구성하여 학습에 사용한다.   
- 병렬 구성한 모델의 학습을 위한 데이터는 학습데이터에서 **복원 추출**로 여러개의 **bootstrap**(sub sample 데이터셋)을 생성하여 학습시킨다.
- 서로 다른 결과를 voting(투표법) 혹은 평균법을 사용하여 개별 모델 결과를 바탕으로 최종 추정치를 얻는다.(aggregation)<br><br>
