## **2. Perceptron(퍼셉트론)**
- 인간의 신경망(뉴런)동작 방식을 모방하여 다수의 입력 신호를 받아 가중합을 계산하고 활성화 함수를 거쳐 0 또는 1의 결과를 출력하는 선형 분류 모델  
- 인공신경망의 가장 기본적인 구성 요소로서 입력, 가중치, 편향, 활성화 함수로 구성
- 직선 하나를 경계로 데이터를 분류하는 선형 분류 모델입니다.  
- AND, OR 연산은 가능하지만, XOR 연산과 같은 비선형 분류는 단일 퍼셉트론으로 불가능

## **3. MLP(Multi layer Perceptron)**
- **`MLP(Multi layer Perceptron)`** 는 사람의 신경을 본따 만든 머신러닝 알고리즘  
- 종속변수가 있어야 하는 **supervised** 모델이며 출력 구분에 따라 **prediction/classification 문제에 모두 사용**이 가능  
- 선형 분류기인 퍼셉트론의 한계를 극복하기 위해 퍼셉트론을 여러 층으로 쌓은 다층 퍼셉트론 구조를 갖는다.  
- 다차원의 비선형성 모델까지 표현이 가능하여 DNN 모델의 기초가 되는 모델

- **input layer** : **설명변수(X)** 를 받아들이는(저장)하는 레이어  
- **hidden layer** : 연산 layer, **input과 output layer를 제외 한 모든 layer**는 hidden layer이다.  
- **output layer** : 출력 layer, 목적식 혹은 **예측/분류에 따라 출력 layer를 다르게** 설정하여 모델링.  
- **node(혹은 뉴런)** : layer에 포함 된 동그라미 하나, **perceptron**  
- **weight(가중치)** : node 와 node간 연결되어 있는 회색 화살표

#### **3.1.2 activation function(활성화함수)**
- 선형 결합 함수의 결과는 여전히 선형, 뉴런의 선형식 결과를 비선형 출력결과로 만들어주는 함수  
- 비선형 활성화 함수를 통해 모델은 복잡한 곡선 형태의 결정 경계를 만들 수 있으며, 현실 세계의 복잡한 데이터를 학습할 수 있습니다.
- 입력 데이터가 특정 임계값을 넘을 때 모델이 해당 정보를 '중요하다'고 판단하여 다음 층으로 전달할지 여부를 결정합니다.
- 출력값을 특정 범위(0 ~ 1 또는 -1 ~ 1)로 제한하여 수치적 안정성을 제공합니다.
- sigmoid, Tanh, ReLu와 같은 함수를 사용하여 출력결과값을 비선형으로 변환한다.
- 현대 딥러닝 모델의 경우 hidden layer에 사용하는 활성화 함수는 특별한 이유가 없으면 ReLU 함수의 변형을 사용하는 것이 일반적

$$\mathbf{z}^{(l)} = \mathbf{W}^{(l)} \mathbf{a}^{(l-1)} + \mathbf{b}^{(l)}$$$$\mathbf{a}^{(l)} = \sigma(\mathbf{z}^{(l)})$$

> $\mathbf{a}^{(l-1)}$: 이전 층(또는 입력층)의 출력 벡터  
> $\mathbf{W}^{(l)}$: 현재 층의 가중치 행렬  
> $\mathbf{b}^{(l)}$: 현재 층의 편향 벡터  
> $\sigma$: 활성화 함수 (예: ReLU, Sigmoid 등)  
> $\mathbf{a}^{(l)}$: 현재 층의 최종 출력 벡터

### **3.3 최적화**
- 결국 cost function이 의미하고자 하는 바는 **비용함수를 최소화 시키는 가중치를 찾고싶다** 이다.
- 일반적인 머신러닝 알고리즘이 가정에 따른 비용함수 최소화를 목적으로 했다면 신경망의 최적화는 **가중치에 따른 비용함수 최소화**를 목적으로 한다.  
- 즉, 학습이 진행되는 스텝마다 가중치를(그래디언트 반대방향으로)업데이트 하면서 최적값을 찾는 과정을 거친다.

#### **3.4 back propagation(역전파)**
- **오차 역전파(Backpropagation)** 는 신경망이 예측값과 실제값 사이의 오차를 줄이기 위해, 출력층에서 입력층 방향으로 거슬러 올라가며 각 층의 가중치($w$)를 얼마나 수정해야 할지 계산하는 알고리즘입니다.
- chain rule을 활용 비용함수로부터 역방향으로 계산한다.

#### **3.5 learning rate(학습률)**
- 신경망 학습에는 학습률이 중요한데 chain rule을 통해 구해지는 업데이트 값에 학습률을 곱해 업데이트 값의 부호 반대방향으로 가중치를 업데이트 한다.  
- 적절한 학습률은 너무 큰 업데이트값을 사용함으로서 모델 학습이 되지 않거나 너무 작은 업데이트 값을 곱해 학습이 더뎌지는 문제를 해결할 수 있다.  
- 가중치 X 학습률 $\eta$ 만큼 업데이트 하면서 최적값을 찾는 과정을 거친다.
# $$W \leftarrow W - \eta{{\partial j}(W)\over{\partial W}}$$

## **4. MLP 실습**
- 딥러닝 학습을 위한 손글씨 이미지 데이터셋으로 MLP모델 실습을 진행합니다.
- 데이터과학 필드에서 이미지 데이터의 클래식이자 이미지 알고리즘에 큰 영향을 준 데이터셋  
- 데이터셋 구성
> - 0~9 까지의 사람 손글씨 이미지데이터 형태의 설명변수와 숫자 레이블 값이 들어가 있는 데이터셋  
> - 28 X 28 데이터의 784개의 변수를 사용하며 총 70000만개의 데이터 샘플을 갖고 있는 데이터

```python
# 필요모듈 import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_openml
from sklearn.neural_network import MLPClassifier
```

```python
# mnist data 로딩
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
```

```python
print(mnist['DESCR'])
```

```python
# 타겟데이터 분할
y = mnist['target'].astype(int)
X = mnist['data'] / 255
```

```python
# 이미지 데이터 확인
plt.imshow(X[0].reshape(28, 28), cmap='gray')
```

```python
pd.Series(y).value_counts()
```

```python
# 학습데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
X_train.shape
```

```python
# 모델 생성
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(
    hidden_layer_sizes=(128, 64),
    activation='relu',
    learning_rate_init=0.001,
    max_iter=20,
    verbose=True,
    random_state=42
)
```

**중요 파라메터**
- hidden_layer_sizes=(100,) : 히든레이어 구조(40, 40, 60) 튜플로 전달  
- activation='relu': 활성함수  
- learning_rate_init=0.001: 학습률  
- max_iter=200 : 학습 반복 횟수  
- verbose=True : iteration에 따른 손실률 확인

데이터에 맞는 히든레이어 구조는 다 다릅니다.  
큰 데이터라도 깊이가 얕은 모델의 성능이 좋을 수 있어 파라메터 서칭(구조 서칭)이 필요합니다.  
일반적으로는 구조가 복잡해질수록 학습률을 줄이면서 반복횟수를 많이 가져가게 됩니다.

이 외 파라메터는 최적화 관련 파라메터인데 현재 가장 좋은 알고리즘이 기본적으로 탑재되어 있으니 그냥 쓰시면 됩니다.


```python
%%time
mlp.fit(X_train, y_train)
```

```python
mlp.loss_curve_
```

```python
# loss curve 시각화
plt.plot(mlp.loss_curve_)
```

```python
# 모델 평가지표 출력
mlp_pred = mlp.predict(X_test)
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, mlp_pred))
```

## **5. DNN 프레임워크 tensorflow를 사용하여 동일한 모델링 시연**

```python
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
import matplotlib.pyplot as plt
import numpy as np
from typing import Any

def build_and_train_tf_mlp(X_train: np.ndarray, y_train: np.ndarray,
                            X_test: np.ndarray, y_test: np.ndarray) -> models.Sequential:
    """
    Build and train a TensorFlow MLP model equivalent to sklearn's MLPClassifier.
    """
    # 1. 모델 정의 (hidden_layer_sizes=(128, 64) 반영)
    model = models.Sequential([
        layers.Input(shape=(784,)),             # 입력층 (MNIST Flatten)
        layers.Dense(128, activation='relu'),   # 은닉층 1
        layers.Dense(64, activation='relu'),    # 은닉층 2
        layers.Dense(10, activation='softmax')  # 출력층 (다중 분류)
    ])

    # 2. 컴파일 (solver='adam' 반영)
    model.compile(
        optimizer=optimizers.Adam(),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    # 3. 모델 학습 (max_iter=20 반영)
    print("TensorFlow 학습 시작...")
    history = model.fit(
        X_train, y_train,
        epochs=20,
        batch_size=200,          # sklearn 기본값은 min(200, n_samples)
        validation_data=(X_test, y_test),
        verbose=1
    )

    # 4. 성능 평가
    test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
    print(f"\n최종 테스트 정확도: {test_acc:.4f}")

    return model, history

def plot_tf_loss_curve(history: Any) -> None:
    """
    Plot the training and validation loss curve using Keras history.
    """
    plt.figure(figsize=(10, 6))
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('TensorFlow MLP Loss Curve')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()

# 실행 예시 (학습 데이터가 정의되어 있어야 함)
tf_model, tf_history = build_and_train_tf_mlp(X_train, y_train, X_test, y_test)
plot_tf_loss_curve(tf_history)
```

- **`DNN`**, **`RNN`**, **`CNN`**, **`LSTM`** 등 딥러닝에 사용되는 베이직한 노드 구조는 4~5년 전 출시 된 구조에서 더는 추가가 되고 있지 않다.  
- 이미지 처리에 **`CNN`**, 자연어 처리에 **`LSTM`** 을 기본구조로 많이 사용
